```python
def make_manager_div(이름, 데프, 타입:str='정부'): # 정부, 직무
    try:
        마스터 = 데프.query("정부 == 'master'")
        담당자 = 데프.query("정부 != 'master'")
        표 = 담당자.pivot(index='부서',columns=타입, values='이름').fillna('-')
        print(이름)
        if len(마스터) :
            print(마스터)
        return 표
    except :
        return '표미생성'
    
```

```python
for 분야,데프 in df.groupby('항목'):
    표 = make_manager_div(분야,데프,'직무')
    print(표)
```
